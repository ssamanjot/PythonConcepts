```python
import numpy as np
```

```python
sampl = np.random.uniform(low=0.5, high=100, size=(1000,2))
sampl.shape
```

```python
Label=np.random.randint(0,2,size=1000)
Label=Label.reshape(1000,1)
```

```python
sampl=np.hstack((sampl,Label))
```

```python
sampl
```

```python
X7=sampl[:700,:]
```

```python
X7.shape
```

```python
Distances=[]
def EuclDist(Xth,Yth,Xsam,Ysam,Label):
    Dist=np.sqrt((Xth-Xsam)**2+(Yth-Ysam)**2)
    Distances.append((Dist,Label))
```

```python

for j in range(0,300):
    Y3=sampl[700+j,:]
    
    #Y3.shape

    X_TestSample=Y3[0]
    Y_TestSample=Y3[1]

    #X_TestSample


    #Y_TestSample

    Label1=Y3[2]
    print("Existing Label :",Label1)


    for i in range(0,len(X7)):
        Xth=X7[i,0]
        Yth=X7[i,1]
        Label=X7[i,2]
        EuclDist(Xth,Yth,X_TestSample,Y_TestSample,Label)


    MiniEucl=min(Distances)

    Distances.remove(MiniEucl)

    MiniEucl1=min(Distances)
    Distances.remove(MiniEucl1)
    MiniEucl2=min(Distances)

    print('3 Minimum distances :',MiniEucl,MiniEucl1,MiniEucl2)
    MinDistances=[]
    MinDistances.append(MiniEucl)
    MinDistances.append(MiniEucl1)
    MinDistances.append(MiniEucl2)

    ZeroTerm=0
    OneTerm=0
    for i in range(0,3):
        if MinDistances[i][1]==0.0:
            ZeroTerm=ZeroTerm+1
        else:
            OneTerm=OneTerm+1
    if ZeroTerm>OneTerm:
        sampl[700+j,2]=0.0
    else:
        sampl[700+j,2]=1.0



    print('Changed Datapoint',sampl[700+j])
    print('-------------------------------------------------------------------------')
```
